## Create data subsets for different sex.

```python
df.SEX.value_counts()
```

## Transposing Data

```python
df_transpose = df.T
df_transpose.head()
```

## Melting Data to long format

```python
df_new = df[["AGE", "SEX", "STEROID"]]
```

```python
df_new_melt = df_new.melt()
df_new_melt.head()
```

```python
df_new.drop_duplicates(subset ="AGE",
                     keep = False, inplace = True)
```

## Casting data to wide format

```python
df_pivot = df_new.pivot('AGE', 'SEX', 'STEROID')
```
